## Oppgave 4 (20 poeng)

Et lån består som regel av et månedlig terminbeløp. Dette beløpet er summen av avdrag (nedbetalingen på lånet) og renter. Vi antar månedlig forrenting i alle oppgavene. Dvs. at det er 12 terminer i hvert år.<br><br>
a) Lag en funksjon som regner ut hvor mye lånet "x" koster deg i renter "t" terminer med årlig rente "r".  Dette er et serielån, så avdragene er like hver måned men renten reduseres i takt med avdragene. 

Formelt, kan vi la totale renteutgifter for hele lånet være summen av renten på gjenværende beløp i alle terminer "t". Dette kan skrives slik   
$\sum_{t=1}^N(x-a*t)*{r/f}$, hvor "a" er avdragene og "f" er antall forrentinger hvert år. F.eks. så vil renten i 6 mnd være
$(x-a*6)*{r/f}$

Formelt, kan vi la perioden "p" definere hvilken mnd vi er i. Dvs, hvis p=15, har det gått 1 år og 3 mnd med terminer. Renteutgiftene for hele lånet vil derfor være summen av renten på gjenværende beløp for alle perioder p. Dette kan skrives slik $\sum_{p=1}^N(x-a*p)*{r/n}$ 

Hint: siden terminbeløpet varierer for hver måned (pga at rentene endres), må alle enkeltperioder summeres. Det kan være nyttige å bruke funksjonen np.arange() til dette.


```python
#versjon 1: regn ut hvor mye lånet koster deg for ulik nedbetalingstid
def serie(x, r, t, n):
    avdrag = x/(t*n)
    rente = (x-avdrag*np.arange(0,t*n))*r/n
    return rente.sum()
```

```python
x = 1000000
t = 3
n = 12
r = 0.03
avdrag = x/(t*n)
rente = (x-avdrag*np.arange(0,t*n))*r/n
rente
```

```python
#versjon 2: regn ut hvor mye lånet koster deg for ulik nedbetalingstid. Her har vi perioder i stedet for år og tid
def serie2(x, r, p):
    avdrag = x/(p)
    rente = (x-avdrag*np.arange(0,p+1))*r/12
    return rente.sum()
```

c) Vis hva det samme lånet koster som annuitetslån, dvs differansen mellom alle terminbeløp og lånebeløp. Annuitetslån gir like terminbeløp hver måned, men renten utgjør en større del av dette beløpet i starten. Terminbeløpet for et annuitetslån er definert ved formelen:
$T = x*\frac{r/n} {(1-(1+(r/n))^{-n*t})}$, hvor x=lånebeløp, r = rente, n = terminer hvert år, t = år





```python
#først må vi finne ut hva termingebyret er 
def termin(x, r, t, n):
    mnd = r/n
    term = t*n
    return x*(mnd/(1-(1+mnd)**-term))
```

```python
termin(1000, 0.03, 1, 12)
```

```python
#kostnaden er summen av terminer minus lånebeløp
def cost(x, r, t, n):
    return termin(x, r, t, n)*n*t-x

```

```python
print(f"Annuitetslånet med 10 års tilbakebetaling koster {round(cost(1000000, 0.03, 10, 12))}kr")
print(f"Annuitetslånet med 20 års tilbakebetaling koster {round(cost(1000000, 0.03, 20, 12))}kr")
print(f"Annuitetslånet med 30 års tilbakebetaling koster {round(cost(1000000, 0.03, 30, 12))}kr")
```

c) Vis hvordan utviklingen i rentekostnader og avdrag på terminer for serielån grafisk. Anta et bankinnskudd x = 1000, renter r=3% og antall perioder p = 36 (det vil si 3 år). Formelt, kan vi la perioden "p" definere hvilken mnd vi er i. Dvs, hvis p=15, har det gått 1 år og 3 mnd med terminer. Renteutgiftene for hele lånet vil derfor være summen av renten på gjenværende beløp for alle perioder p. Dette kan skrives slik $\sum_{p=1}^N(x-a*p)*{r/n}$ 

<br> Bruk stackplot til dette. <br> <br>

<b>Hint1:</b> Siden avdragene er like for alle måneder, kan det være lurt å definere det månedlige avdraget som en liste og gange det med antall perioder. </br>
<b>Hint2:</b> Det kan være enklere å definere tid på enkeltmåneder når man skal vise det grafisk (ref, kommentar om periode p ovenfor) 


```python
#samme som tidligere funksjon, men her tar vi ikke summen
def interest_rate_monthly_serie(x, r, term):
    avdrag = x/(term)
    rente = (x-avdrag*np.arange(0,term+1))*r/12
    return rente
```

```python
#tester funksjonen
interest_rate_monthly_serie(1000, 0.03, 36)
```

```python
#lager funksjon for downpayment
def down_payment(x, term):
    return x/term
```

```python
y = np.arange(1, 37, 1)
x = 36
x1 = [down_payment(1000,36)]*x
x2 = interest_rate_monthly_serie(1000, 0.2, 35)

fig, ax = plt.subplots()
ax.stackplot(y, x1, x2, labels=["avdrag", "renter"])
ax.legend(loc='upper left')
ax.set_title('Lån')
ax.set_xlabel('Antall mnd')
ax.set_ylabel('Termingbeløp')

plt.show()
```
