### Oppgave 3 (20 poeng)

En bedrift produserer biler. Produktfunksjonen til bedriften defineres slik  $f(L, a, R) = 2RL^{a}$, hvor:
- `L` er arbeidskraft, 
- `a` er produktiviteten til arbeiderne og 
- `R` er antall robotmaskiner

a) Lag en formel for produktfunksjonen til bedriften og plot den grafisk med ulike verdier av `L` på x-aksen. Anta `a=0.6` og `R=2`

```python
#setter inn funksjonen og lager X-aksen som en linspace
a=0.6
R=2
#definerer produktfunksjonen funkjsonen
def produktfunksjon(L, a, R):
    return 2*R*L**a

L = np.linspace(0, 10000, 10000) #lager en linspace (L) for x
#lager plot
plt.plot(L, produktfunksjon(L, a, R))
plt.ylabel("Biler solgt")
plt.xlabel("Arbeidere")
plt.legend(['Produktfunksjon'])
```

b) anta at profittfunksjonen til denne bedriften er ${profit = f(L, a, R)p-wL-cR-K}$, hvor 
- `w` er månedslønnen til arbeiderne, 
- `c` er kostnaden for robotmaskinene 
- `K` er faste kostnader
- `p` er utsalgsprisen på bilene. 

Anta `a=0.6`,`R=6`, `p=300 000`, `w=100 000`, `c= 1 000 000` og `K=90 000 000`. Plot profittfunksjonen figurativt for antall arbeidere (`L`) mellom 0 og 10 000. Vis profitten i millioner (dvs at du må dele på 1 000 000)

```python
#setter inn verdier og profittfunksjonen som en funksjon av produktfunksjonen
a=0.6
R=6
p=300000
w=100000
c=1000000
K=90000000

#definerer prodittfunksjonen funksjonen 
def profittfunksjon(L,a,R,p,w,c,K):
    return ((produktfunksjon(L, a, R))*p-w*L-c*R-K)/1000000



#lager plot

plt.plot(L, profittfunksjon(L,a,R,p,w,c,K))
plt.ylabel("Profitt (millioner kr)")
plt.xlabel("Arbeidere")
plt.legend(['Profittfunksjon'])
```

c) Plot profitttfunksjonen for antall robostmaskiner `R=[3, 6, 9]` i samme plot (dvs at tre profittfunksjoner vises sammen). Bruk av "for loops" for å gjøre dette belønnes

```python
#plotter samme plottet for de forskjellige verdiene av R ved bruk av en for løkke
for R in [3,6,9]: 
    plt.plot(L,profittfunksjon(L,a,R,p,w,c,K), label=R)
plt.ylabel("Profitt (millioner kr)")
plt.xlabel("Arbeidere")
plt.legend()
plt.show()
```

d) finn profittmaksimum og optimal antall arbeidere ved hjelp av derivasjon med samme forutsetninger som i (1b). Bruk `sympy`-pakken til dette 

```python
#finner formel for optimalt antall arbeidere

#definerer symboler som sympy symboler
L,a,R,p,w,c,K=sp.symbols("L a R p w c K")


#finner optimalt antall arbeidere utifra L

#finner den derivierte med hensyn i L
d_profitt=sp.diff(profittfunksjon(L,a,R,p,w,c,K),L)


#setter den deriverte til 0 
foc=sp.Eq(d_profitt,0)

#løser likningen og finner dermed profittmaksimum

L_max=solve(foc,L)[0]
L_max
```

```python
#bruker likningen vår for å finne numerisk optimale arbeidere ved å sette inn de riktige verdiene
num_dict={a:0.6,R:6,p:300000,w:100000}
L_max = L_max.subs(num_dict)

L_max
```

```python
#finner Y verdien til dette punktet på grafen


#legger til L_max verdien vår i funksjonen for å finne profittmaskimums punktet
num_dict={a:0.6,R:6,p:300000,w:100000, c:1000000, K:90000000}

#legger L_max til i orbok
num_dict[L]=L_max.subs(num_dict)

profit_max_num= profittfunksjon(L,a,R,p,w,c,K).subs(num_dict)
profit_max_num
```
